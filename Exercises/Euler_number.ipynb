```python
# import libararies/backgages/ frame_work/
import matplotlib.pyplot as plt
```

```python
# create function
def estimate_e_number(n):
    '''This is funcion to estimate the euler number
        n : the input number

        return the estimated e number
        
    '''
    if n <=0:
        print ('Wrong input in function estimate_e_number')

    else:    
        result = (1+(1/n))**n
        
    return result

```

```python
# test function
e_number=estimate_e_number(1)
# print(e_number)
```

```python
# plot a finger
n=list(range(0,1000,5))
# print(n)
list_e=[]
for i in n:
    if i<=0:
        list_e.append(0)
        continue
    else:
        e_number=estimate_e_number(i)
        #print(e_number)
        list_e.append(e_number)
        #print(list_e)

plt.plot(n,list_e,"green")
#plt.show()
```
