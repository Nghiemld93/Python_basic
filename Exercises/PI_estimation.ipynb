```python
def PI_estimate(n): # Calculus pi number by Gregory-Leiniz Series
    pi=0
    for i in range(n):
        pi+=(-1)**i/(2*i+1)


    return pi*4  

 
def PI_estimate_Nis(n): # Calculus pi number by Nilakantha Series ( more accuracy than Gregory-Leiniz)
    pi=0
    for i in range(n):
        pi+=(-1)**i/((2*i+2)*(2*i+3)*(2*i+4))
    return (pi*4)+3   
```

```python
#test PI_estimate() function:
n=PI_estimate(1000)
```

```python
# plot a finger
import matplotlib.pyplot as plt
n=list(range(0,1000,1))
# print(n)
list_pi=[]
for i in n:
        e_number=PI_estimate_Nis(i)
        #print(e_number)
        list_pi.append(e_number)
        #print(list_pi)

plt.plot(n,list_pi,"green")
#plt.show()
```
